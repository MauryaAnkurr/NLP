```python

#The nlp function from spacy automatically takes raw text and performs
#a series of operatios to tag parse and describe the text data.
import spacy
nlp =spacy.load('en_core_web_sm')
```

```python
myfile = open("TextDocuments.txt")
text = myfile.read()
myfile.close()
```

```python
doc = nlp(text)
```

```python
for token in doc:
    print(token.text,"++",token.pos_)
    print("++++++++++++++")
    
```
